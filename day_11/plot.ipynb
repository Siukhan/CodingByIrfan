```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset

dataset = sns.load_dataset('titanic')
dataset
```

```python
dataset.info()
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x ="survived", hue="sex", data=dataset)
```

```python
# Scatter Plot
plt.scatter(data=dataset, x="age", y = "fare", color = "red");
```

```python
sns.scatterplot(data=dataset, x="sex", y = "fare", hue = "pclass");
```

```python
sns.barplot(data=dataset, x="sex", y = "fare", hue = "pclass");
```

```python
dataset.columns
```

```python
sns.pairplot(dataset);
```

```python
dataset.hist(bins=10, figsize = (9,7), grid=False);
```

```python
sns.boxenplot(data=dataset, x="sex", y = "fare", hue = "class");
```

```python
# How to change the x and y axis titles
```

```python
g = sns.FacetGrid(dataset, hue="survived", col="pclass", margin_titles=True,
              palette={1:"seagreen", 0:"gray"})
g=g.map(plt.scatter, "fare", "age", edgecolor="b").add_legend();
```

```python
g = sns.FacetGrid(dataset, hue="survived", col="who", margin_titles=True,
              palette={1:"seagreen", 0:"gray"})
g=g.map(plt.scatter, "fare", "age", edgecolor="b").add_legend();
```

```python
sns.stripplot(data=dataset, x="sex", y = "fare", hue = "class");
```

```python
# Filter out non-numeric columns
numeric_columns = dataset.select_dtypes(include=[float, int]).columns

# Calculate the correlation matrix
corr = dataset[numeric_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, linewidths=0.5, fmt=".2f")
plt.show()
```

```python
sns.violinplot(data=dataset, x="sex", y = "fare", hue = "class");
```
